## Introduction:

The aim of analysing the given data set is to understand the patterns of how the prices and sizes of the houses are varying with the income levels of individuals,the median values of all variables, how far the housing communities are from the ocean with their longitudinal and latitudinal positions.




## Data Description:

1) LONGITUDES : These are vertical imaginary lines drawn on the globe basically to locate every piece of land on earth. In this data set, based on londitudinal and latitudinal lines, locations of housing communities are given.

2) LATITUDES : These are horizontal imaginary lines drawn on the globe basically to locate every piece of land on earth. In this data set, based on londitudinal and latitudinal lines, locations of housing communities are given.

3) HOUSING MEDIAN AGE : Shows the Median ages of houses.

4) TOTAL ROOMS : Based on the longitudes and latitudinal position, total number of rooms in each housing community.

5) TOTAL BEDROOMS : Among all the rooms, only the count of bed rooms is shown in this column exclusively.

6) POPULATION : Total number of people living in that particular location.

7) HOUSEHOLDS : Total number of individual families living in that location.

8) MEDIAN INCOME : Median of all the earnings from the population living in particular location.

9) MEDIAN HOUSE VALUES : The median value of houses in each housing community based on their latitudinal and longitudinal location.

10) OCEAN PROXIMITY : Tells us about the housing communities' nearness to the Ocean (Near Bay, Near Ocean, Inland)

1. What is the average median income of the data set and check the distribution of data using appropriate plots.
Please explain the distribution of the plot.

```python
avg_med_income = df_housing['median_income'].mean()  # average median income of the data set using '.mean()' function
```

```python
print("The average median income of the data set is:",avg_med_income)
```

```python
plt.hist(df_housing['median_income'], color="orange", edgecolor="white",bins=40)   # hist function is used to plot the histogram for median_income variable
plt.xlabel("Median Income")       # xlabel, ylabel are for naming the x and y axis
plt.ylabel("No.of occurences")
plt.title("Distribution of Median Income using a Histogram")   # title is to give a title to the histogram
plt.axvline(x = df_housing.median_income.mean(),color="darkgreen",alpha=1,label="Mean")
plt.axvline(x = df_housing.median_income.median(),color="darkred",alpha=1,label="Median")
plt.legend()
plt.show()                        # show() is used to display the graph
```

```python
plt.boxplot(df_housing["median_income"], vert=False)   # .boxplot function is used to draw a boxplot for given variable
plt.title("Box Plot for Median Income")                # title is to give a title to the boxplot
plt.text(y=0.85,x=df_housing["median_income"].median(),s='Median')
plt.show()
```

```python
Median = df_housing['median_income'].median()
q1 = np.quantile(df_housing["median_income"],0.25)
q3 = np.quantile(df_housing["median_income"],0.75)
IQR = q3-q1
Lower_Whisker = q1 - (1.5 * IQR)
Upper_Whisker = q3 + (1.5 * IQR)
print("Median:",Median,"\nI Quartile:",q1,"\nIII Quartile:",q3,"\nInter Quartile Range:",IQR,"\nLower Whisker:",Lower_Whisker,"\nUpper Whisker:",Upper_Whisker)
```

```python
# code for retrieving all the outliers

Outliers = []

for i in df_housing["median_income"]:
  if((i < Lower_Whisker) or (i > Upper_Whisker)):
    Outliers.append(i)

print("Outliers from the variable Median_Income are following:\n\n",Outliers)
```

```python
'''From the box plot we can see that the mean is very close to median(from calculations mean is 3.87,median is 3.53) ie., the data is right skewed,
the outliers start from value 8'''
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
df_housing.shape    # .shape function returns the total no.of rows and columns of data set
```

```python
New_housing_df = df_housing.copy()         # deep copying the original dataset to a new variable
New_housing_df
```

```python
New_housing_df["total_bedroom_size"]=""    # creating a new empty column to update the calculated values
df_housing.size

i=0                                        # a code is written to categorise the data based on no.of bedrooms as large,medium,small
while(i<len(df_housing)):
  if(df_housing["total_bedrooms"].loc[i] <= 10):
    New_housing_df["total_bedroom_size"].loc[i] = "Small"
  elif(11 <= df_housing["total_bedrooms"].loc[i] <= 1000):
    New_housing_df["total_bedroom_size"].loc[i] = "Medium"
  else:
    New_housing_df["total_bedroom_size"].loc[i] = "Large"
  i=i+1
```
